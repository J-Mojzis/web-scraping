```python
# Calculate correlation matrix
correlation_matrix = clean_data.iloc[:, 1:].corr()

# Extract the top 15 positive and 15 negative correlations
correlations = correlation_matrix.unstack().reset_index()
correlations.columns = ['Stock1', 'Stock2', 'Correlation']

# Remove self-correlations
correlations = correlations[correlations['Stock1'] != correlations['Stock2']]

# Sort by absolute correlation
correlations['AbsoluteCorrelation'] = correlations['Correlation'].abs()
correlations_sorted = correlations.sort_values(by='AbsoluteCorrelation', ascending=False)

# Get the top 15 most positive and 15 most negative correlations
top_15_positive = correlations_sorted[correlations_sorted['Correlation'] > 0].head(15)
top_15_negative = correlations_sorted[correlations_sorted['Correlation'] < 0].tail(15)

# Identify leading stocks (stocks influencing others)
# Calculate lagged correlations (with 1-day delay)
lags = 2
lagged_data = clean_data.copy()
lagged_data.iloc[:, 1:] = lagged_data.iloc[:, 1:].shift(lags)

# Compute lagged correlation matrix
lagged_correlation_matrix = clean_data.iloc[:, 1:].corrwith(lagged_data.iloc[:, 1:], axis=0)

# Sort stocks by their influence (sum of lagged correlations)
leading_stocks = lagged_correlation_matrix.abs().sort_values(ascending=False).head(10)

# Display results
print("Top 15 Most Positive Correlations:")
print(top_15_positive)

print("\nTop 15 Most Negative Correlations:")
print(top_15_negative)

print("\nTop 10 Leading Stocks (Most Influential):")
print(leading_stocks)

# Visualizing leading stocks
plt.figure(figsize=(10, 5))
plt.bar(leading_stocks.index, leading_stocks.values, color='blue')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Lagged Correlation (Absolute)')
plt.title('Top 10 Leading Stocks (Most Influential)')
plt.tight_layout()
plt.show()
```

```python
# Define the lag
lag_days = 2

# Shift the data by 2 days to calculate lagged influence
lagged_data = clean_data.copy()
lagged_data.iloc[:, 1:] = lagged_data.iloc[:, 1:].shift(lag_days)

# Get the top 3 leading stocks
top_3_leading_stocks = leading_stocks.index[:3]

# Dictionary to store influenced stocks for each leading stock
influenced_stocks = {}

for leading_stock in top_3_leading_stocks:
    # Calculate lagged correlations for the current leading stock
    lagged_correlations = clean_data.iloc[:, 1:].corrwith(
        lagged_data[leading_stock], axis=0
    )
    
    # Drop the leading stock itself and sort by descending correlation
    lagged_correlations = lagged_correlations.drop(leading_stock, errors='ignore')
    top_3_influences = lagged_correlations.nlargest(3)
    
    # Store the results
    influenced_stocks[leading_stock] = top_3_influences

    
# Display results
for leading_stock, influences in influenced_stocks.items():
    print(f"\nTop 3 Stocks Influenced by {leading_stock} (2-day lag):")
    print(influences)
    # Plot the top 3 influenced stocks
    plt.figure(figsize=(8, 5))
    plt.bar(influences.index, influences.values, color='skyblue')
    plt.xlabel('Influenced Stocks')
    plt.ylabel('Lagged Correlation (2-day)')
    plt.title(f'Top 3 Stocks Influenced by {leading_stock} (2-day lag)')
    plt.tight_layout()
    plt.show()
```
